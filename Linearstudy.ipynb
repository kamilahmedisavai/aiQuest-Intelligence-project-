```python
import pandas as pd 
import numpy as np
```

```python
#dataset 
data ={ 
    'Weight' : [2,4,5,3,6,5,7],
    'Price' : [35,60,20,50,50,55,60]
    }
df = pd.DataFrame(data)
```

```python
df
```

```python
# find M and C
```

```python
X = df['Weight']
y = df['Price']
n = len(X)

sum_x = np.sum(X)
sum_y = np.sum(y)
sum_xy = np.sum(X * y)
sum_x2 = np.sum(X ** 2)
```

```python
M = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
C = (sum_y - M * sum_x) / n
```

```python
pred_6_50 = M * 6.5 + C
```

```python
df['Predicted'] = M * X + C
df['Residual'] = df['Price'] - df['Predicted']
```

```python
mse = np.mean(df['Residual'] ** 2)
mae = np.mean(np.abs(df['Residual']))
```

```python
print(f"Slope (M): {M:.4f}")
print(f"Intercept (C): {C:.4f}")
print(f"Predicted Price for 6.50: {pred_6_50:.4f}")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
```
